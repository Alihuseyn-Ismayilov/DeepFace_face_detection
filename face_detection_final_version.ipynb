```python

import tkinter
from tkinter import *
from PIL import Image, ImageTk
import cv2
import datetime
import threading
import cv2
from deepface import DeepFace
from PIL import Image
import PIL
import numpy as np
import tkinter.font as font

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)


cap.set(cv2.CAP_PROP_FRAME_WIDTH,1000)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,1000)

root = Tk()

def on_closing():
    cap.release()
    cv2.destroyAllWindows()
    root.destroy()
    
root.protocol("WM_DELETE_WINDOW", on_closing)




label1 = tkinter.Label(root,image=None)#test

label1.grid(row=1,column=0)


verified_picture_label = tkinter.Label(root,image=None)
verified_picture_label.grid(row=1,column=1)


alihuseyn = cv2.imread('resume_photo.jpg')
bill = cv2.imread('bill4.jpg')
elon = cv2.imread('elon2.jpg')

known_faces = {
    
    'Alihuseyn' : alihuseyn,
    'Bill Gates': bill,
    'Elon Musk' : elon
    
}


count = 3

right_label = tkinter.Label(root,text='',fg='red')
myFont = font.Font(family='Helvetica', size=15, weight='bold')
right_label['font'] = myFont
right_label.grid(row=2,column=1)


def change_pic():
    global verified_picture,count
    
    
    right_pic_tk = frame_Image.copy()
    
    
    frame_cv = np.array(right_pic_tk)
    frame_cv = cv2.cvtColor(frame_cv, cv2.COLOR_RGB2BGR)
    
    
    right_label['text'] = 'Verification Process'
    right_label.grid(row=2,column=1)
    root.update()
    
    flag = False
    
    for name,verify_image in known_faces.items():
        result = DeepFace.verify(frame_cv,verify_image,model_name='Facenet'\
                                     ,detector_backend='retinaface',enforce_detection=False) 
        bounding_box = list(result['facial_areas']['img1'].values())
        if(result['verified'] == True):
            x, y, w, h = int(bounding_box[0]), int(bounding_box[1]), int(bounding_box[2])\
            , int(bounding_box[3])
            cv2.rectangle(frame_cv, (x, y), (x + w, y + h), (0,255,0), 2)

            cv2.putText(frame_cv, name, (x, y - 10) , cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
            
            flag = True
            count = 3
    if(flag):
        right_label['text'] = 'Verified'
        right_label.grid(row=2,column=1)
    else:
        count -= 1
        if(count == 0):
            right_label['text'] = f'"Sorry, all attempts used. Please contact support for help'
            button['state'] = 'disabled'
        else:
            right_label['text'] = f'Invalid Password.Please try again.You have {count} more attempts remaining.'
        right_label.grid(row=2,column=1)
    
    verified_picture = cv2.cvtColor(frame_cv,cv2.COLOR_BGR2RGB)
    verified_picture = Image.fromarray(verified_picture)
    verified_picture = ImageTk.PhotoImage(verified_picture)
    verified_picture_label['image'] = verified_picture
    verified_picture_label.grid(row=1,column=1)
    
    
    
    
    
    
    
    

button = tkinter.Button(root, text ="Verify", command = change_pic,fg='white',bg='blue')
myFont = font.Font(family='Helvetica', size=15, weight='bold')
button.grid(row=2,column=0,ipadx=30,ipady=5,pady=3)
button['font'] = myFont



while True:
    ret,frame = cap.read()
    if(ret):
        
        
        
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        frame_Image = Image.fromarray(frame)

        frame_Tk = ImageTk.PhotoImage(frame_Image)
                
        label1['image'] = frame_Tk
        label1.grid(row=1,column=0)

        
    else:
        print('could not get video frame')
        break
            
    root.update()

    
```
